```python
#alternative rows and alternative column
alt_rows=[]
for i in range(0,mtcars.shape[0]):
    if i%2!=0:
        odd=i
        alt_rows.append(odd)
        
alt_columns=[]
for j in range(0,mtcars.shape[1]):
    if j%2==0:
        even=j
        alt_columns.append(even)
```

```python
print(f'the alternate number of rows :    {alt_rows}')
print(f'the alternate number of columns : {alt_columns}')
```

```python
mtcars.iloc[alt_rows,alt_columns]
```

```python
#find row with Mazda RX4 Wag and columns cyl, am
mtcars.loc['Mazda RX4',['cyl','am']]
```

```python
#find row between Merc 280 and Volvo 142E Mazda RX4 Wag and columns cyl, am
mtcars.loc['Merc 280':'Volvo 142E',['cyl','am']]
```

```python
# mpg > 23 or wt < 2
mtcars[(mtcars.mpg>23) | (mtcars.wt < 2)]
```

```python
#using lambda for above
lamb=lambda mtcars:mtcars[(mtcars.mpg>23) | (mtcars.wt < 2)]
lamb(mtcars)
```
